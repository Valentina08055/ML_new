График функции (sin(2x+1))^5

```python
import numpy as np
import matplotlib.pyplot as plt 
  # векторы абсцисс и ординат
X = np.linspace(-5, 5)
Y = np.linspace(-1,1)
Y = np.sin(2*X+1)**5
  # создание графика и его отображение
plt.plot(X,Y) 
plt.show()
```
