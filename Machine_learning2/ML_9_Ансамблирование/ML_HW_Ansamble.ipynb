```python
# создаем матрицу корреляций
corr_matrix = data.corr().abs()

# Выбираем верхний треугольник матрицы
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Ищем индексы колонок с корреляцией большей чем 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

print(to_drop)
```

```python
data_categ  = data.select_dtypes(include=['object'])
data_categ
```

```python
data_categ = pd.get_dummies(data_categ)
data_categ = data_categ.iloc[:,1:156]
data_categ
```

```python
data_new = pd.concat([data_num, data_categ],  axis = 1)
```

```python
data_new
```

Нормализуем числовые значения

```python
# Выделим колонки в отдельный список
headers = list(data_new.columns.values)
```

```python
# Нормализуем параметры
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1,1))

np_scaled = min_max_scaler.fit_transform(data_new)
df_norm = pd.DataFrame(np_scaled)
df_norm.columns = headers
df_norm
```

```python
X = df_norm
```

```python
y = data['SalePrice']
```

делим данные на тестовые и тренировочные

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2)
```

используем модель Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)

# обучаем модель
model.fit(X_train,y_train)
```

```python
headers = list(X_train.columns.values)

feature_imp = pd.Series(model.feature_importances_,index=headers).sort_values(ascending=False)

f, ax = plt.subplots(figsize=(20, 40))
sns.barplot(x=feature_imp, y=feature_imp.index)

plt.xlabel('Важность атрибутов')
plt.ylabel('Атрибуты')
plt.title("Наиболее важные атрибуты")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import classification_report
model_pred = model.predict(X_test)

print(classification_report(y_test, model_pred))

model.score(X_test, y_test)
```
