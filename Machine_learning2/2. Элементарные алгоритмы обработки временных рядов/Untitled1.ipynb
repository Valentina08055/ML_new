```python
import pandas as pd
import numpy as np
```

```python
sales_of_company_x = pd.read_csv("D:\Python3\Machine_learning2/monthly-sales-of-company-x-jan-6.csv")
```

```python
sales_of_company_x.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sales = sales_of_company_x.Count
```

```python
with plt.style.context('bmh'):
    plt.plot(sales, color = 'green')
```

```python
with plt.style.context('bmh'):
    for i in sales:
        rolling_mean = sales.rolling(window=7).mean()
    plt.figure(figsize=(14, 8))
    plt.plot(rolling_mean, color = 'red')
    plt.plot(sales, color = 'green')
```
