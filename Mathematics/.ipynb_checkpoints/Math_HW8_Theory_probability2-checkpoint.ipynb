Задание 1
Получить интуицию, на что влияют параметры распределений. 
Построить по 5 функций плотностей с разными параметрами для каждого распределения: 
нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

%matplotlib inline
```

#### Нормальное распределение

```python
mu = 2.0
mu2 = 2.0
mu3 = 2.0
mu4 = 3.0
mu5 = 1.0
sigma = 0.5
sigma2 = 0.8
sigma3 = 0.3
sigma4 = 0.5
sigma5 = 0.5

# зададим нормально распределенную случайную величину
norm_rv = sts.norm(loc=mu, scale=sigma)
norm_rv2 = sts.norm(loc=mu2, scale=sigma2)
norm_rv3 = sts.norm(loc=mu3, scale=sigma3)
norm_rv4 = sts.norm(loc=mu4, scale=sigma4)
norm_rv5 = sts.norm(loc=mu5, scale=sigma5)

# сгенерируем 10 значений
norm_rv.rvs(size=10)
norm_rv2.rvs(size=10)
norm_rv3.rvs(size=10)
norm_rv4.rvs(size=10)
norm_rv5.rvs(size=10)

# Параметр loc задаёт $\mu$, scale — среднеквадратичное отклонение $\sigma$,
# size — размер выборки. Имя параметра size при вызове функции rvs можно не писать.
```

```python
x = np.linspace(0,4,100)

pdf = norm_rv.pdf(x)
pdf2 = norm_rv2.pdf(x)
pdf3 = norm_rv3.pdf(x)
pdf4 = norm_rv4.pdf(x)
pdf5 = norm_rv5.pdf(x)

plt.plot(x, pdf)
plt.plot(x, pdf2)
plt.plot(x, pdf3)
plt.plot(x, pdf4)
plt.plot(x, pdf5)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

#### Экспоненциальное распределение

```python
lam = 0.1
expon_rv = sts.expon(lam)
expon_rv.rvs(10)
```

```python
x = np.linspace(0,5,100)
x2 = np.linspace(0,20,100)
x3 = np.linspace(0,2,100)
x4 = np.linspace(0,10,100)
x5 = np.linspace(1,5,100)

cdf = expon_rv.cdf(x)
cdf = expon_rv.cdf(x2)
cdf = expon_rv.cdf(x3)
cdf = expon_rv.cdf(x4)
cdf = expon_rv.cdf(x5)

plt.plot(x, cdf)
plt.plot(x2, cdf)
plt.plot(x3, cdf)
plt.plot(x4, cdf)
plt.plot(x5, cdf)

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```

#### Распределение Стьюдента

```python
n = 10

t_rv = sts.t(n)

t_rv.rvs(10)
```

```python
x1 = np.linspace(-5,5,100)
x2 = np.linspace(-5,10,100)
x3 = np.linspace(-1,5,100)
x4 = np.linspace(-15,3,100)
x5 = np.linspace(-8,5,100)

cdf = t_rv.cdf(x1)
cdf = t_rv.cdf(x2)
cdf = t_rv.cdf(x3)
cdf = t_rv.cdf(x4)
cdf = t_rv.cdf(x5)

plt.plot(x1, cdf)
plt.plot(x2, cdf)
plt.plot(x3, cdf)
plt.plot(x4, cdf)
plt.plot(x5, cdf)


plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```
