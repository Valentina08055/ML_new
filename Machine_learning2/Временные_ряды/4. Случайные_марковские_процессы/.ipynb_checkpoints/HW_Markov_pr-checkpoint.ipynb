Скачайте текст Войны и мира, оставив только русские буквы и пробелы. Постройте матрицу переходных вероятностей. Посчитайте вероятность встретить строку из 4-х символов, которая начинается на п и заканчивается на р. 
Используя матрицу переходных вероятностей, сгенерируйте слово, начинающееся с буквы "Б" и имеющее длину 8 символов.

```python
with open(r'D:\Python3\Machine_learning2\Временные_ряды\tom_1_war_and_peace.txt' ,'r', encoding='UTF-8' ) as f:
    text = f.read().lower()

text = re.sub('[^а-я]', " ", text)
```

```python
text = ' '.join(map(str.strip,text.split()))
```

```python
# text
```

```python
RUSSIAN = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя '
POS = {l: i for i, l in enumerate(RUSSIAN)}
# POS
```

```python
probabilities = np.zeros((len(RUSSIAN), len(RUSSIAN)))
# probabilities
```

```python
for i, j in zip(text[:-1], text[1:]):
    probabilities[POS[i], POS[j]] += 1
```

```python
zzz = zip(text[:-1], text[1:])
```

```python
for cp, cn in zip(text[:-1], text[1:]):
    probabilities[POS[cp], POS[cn]] += 1
```

```python
probabilities = (probabilities / probabilities.sum(axis=1)[:, None])
```

```python
df = pd.DataFrame(probabilities, index=(c for c in RUSSIAN), columns=(c for c in RUSSIAN))
```

```python
df = df.dropna()
df = df.drop(['ё'], axis =1)
```

```python
df
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(df)
plt.show()
```

```python
start = POS['п']
end = POS['р']
prior = 1. / len(RUSSIAN)
for i in RUSSIAN:
    for j in RUSSIAN:
        proba = prior * probabilities[start, POS[i]] * probabilities[POS[i], POS[j]] * probabilities[POS[j], end]
        if proba > 0:
            print(''.join(['п', i, j, 'р']), proba)
```

```python
from hmmlearn.hmm import GaussianHMM
```

```python
series = pd.read_csv("D:\Python3\Machine_learning2\Временные_ряды/monthly-sales-of-company-x-jan-6.csv")["Count"]
```

```python
values = series.values.reshape(-1,1)
```

```python
for n in range(1, 14):
    model = GaussianHMM(n_components=n,
                        covariance_type="diag",
                        n_iter=1000)
    model.fit(values)
    print (n, model.score(values))
```

```python
model = GaussianHMM(n_components=8,
                        covariance_type="diag",
                        n_iter=1000)
model.fit(values)
labels = model.predict(values)
means = np.zeros_like(values)
for i in range(model.n_components):
    means[labels == i] = model.means_[i]
    
plt.figure(figsize=(12, 6))
plt.plot(values)
plt.plot(means, linewidth=3)
```
