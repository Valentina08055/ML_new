```python
def desigion_tree(data):
    if data['Goal'] == "иду за хлебом":
        return "no"
    if data['Goal'] == "иду за платьем":
        if data['Color'] == "красный":
            return "yes"
        if data['Color'] == "черный":
            return "yes"
        if data['Color'] == "желтый":
            return "no"
    if data['Goal'] == "иду за сумкой":
        if data['Mood'] == "хорошее":
            return "yes"
        if data['Mood'] == "не очень":
            return "no"
```

```python
data['Desigion'] = data.apply(desigion_tree, axis = 1)
data
```
